# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score,confusion_matrix
```

```python
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
ytest_predict=model.predict(x_test)
ytrain_predict=model.predict(x_train)
```

```python
print('Accuracy Test:',accuracy_score(y_test,ytest_predict))
print('Accuracy Train:',accuracy_score(y_train,ytrain_predict))
      
```

```python
precision=precision_score(y_test,ytest_predict,average='macro')
recall=recall_score(y_test,ytest_predict,average='macro') #sensibildiad
F1=f1_score(y_test,ytest_predict,average='macro')
```

```python
print('Presicion:',precision)
print('Recall:',recall)
print('F1_score:',F1)
```

```python
test_.append(accuracy_score(y_test,ytest_predict))
train_.append(accuracy_score(y_train,ytrain_predict))
precision_.append(precision)
recall_.append(recall)
f1_.append(F1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

```

```python
x_notScaler=df.iloc[:,1:]
```

```python
y_pip=encoder.fit_transform(y1)
y_pip
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_notScaler, y_pip, test_size=0.2, shuffle=True, random_state=42)
```

```python


pipeline = Pipeline([
    ('scaler', MinMaxScaler()),  
    ('logistic_regression', LogisticRegression())  
])


pipeline.fit(x_train, y_train)


ytest_predict_lr = pipeline.predict(x_test)
ytrain_predict_lr = pipeline.predict(x_train)
```

```python
print('Accuracy Test:',accuracy_score(y_test,ytest_predict_lr))
print('Accuracy Train:',accuracy_score(y_train,ytrain_predict_lr))
```
